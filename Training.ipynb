```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from skimage.feature import hog
from sklearn.model_selection import train_test_split
from sklearn import datasets, linear_model
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.model_selection import KFold

# LOAD FILE
Xtr=np.loadtxt("TrainData.csv")
Xts=np.loadtxt("TestData.csv")
Ytr=np.loadtxt("TrainLabels.csv")

# READING IMAGES INTO LISTS
def reshape_images(Xtr):
    list_img = []
    for row in Xtr:
        list_img.append(row.reshape([28,28]))
    return list_img

# HOG EXTRACTED FEATURES
def extract_features(reshaped_images):
    ext_feat = []
    for features in reshaped_images:
        ext_feat.append(hog(features, orientations=8, pixels_per_cell=(8, 8), cells_per_block=(2, 2), visualize=False, multichannel=False))
    return ext_feat

# SELECTION OF FEATURES FROM IMAGE VECTORS
ext_features = extract_features(reshape_images(Xtr))

ext_features=np.array(ext_features)
print (ext_features[1])

# VARIABLES DECLARATIONS
accs=[]
accs1=[]
models=[]
models1=[]
maxind=0
maxind1=0
maxacc1=0
maxacc=0
i=0

# NUMBER OF KFOLDS
kf = KFold(n_splits=5)
for train_index, test_index in kf.split(ext_features):
    regressor = RandomForestClassifier(n_estimators=5, random_state=0)
    regressor.fit(ext_features[train_index], Ytr[train_index].ravel())
    log_reg = LogisticRegression()
    log_reg.fit(ext_features[train_index], Ytr[train_index].ravel())
    y_pred = regressor.predict(ext_features[test_index])
    y_pred1 = log_reg.predict(ext_features[test_index])
    acc=np.sum(y_pred==Ytr[test_index].ravel())/float(test_index.shape[0])
    acc1=np.sum(y_pred1==Ytr[test_index].ravel())/float(test_index.shape[0])
    accs.append(acc)
    accs1.append(acc1)
    models1.append(log_reg)
    models.append(regressor)
    
    if(acc>maxacc):
        maxacc=acc
        maxind=i
    if(acc1>maxacc1):
        maxacc1=acc1
        maxind1=i
    i+=1
    
print(accs[maxind],accs1[maxind1])
regres=models[maxind]
log_regs=models1[maxind1]

import pickle
with open('myModel.pkl', 'wb') as f:
    pickle.dump((regres,log_regs), f)
```
